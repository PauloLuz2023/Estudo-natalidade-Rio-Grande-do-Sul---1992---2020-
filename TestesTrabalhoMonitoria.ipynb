Calcular taxa de crescimento a partir do coeficiente angular da reta do gráfico normalizado -> Y = aX + b

EQUAÇÃO PARA OBTER O COEFICIENTE ANGULAR

â = n*Soma(x*y) - Soma(x)*Soma(y) / nSoma(x^2) - (Soma(x))^2

â = coeficiente angular

n = numero de elementos de y

x = lista com index dos anos: index do ano -> 2001 = 1, 2002 = 2, etc.

y = lista com nascimentos com vida por ano

```python
import numpy as np
```

```python
def calcula_taxa_de_crescimento(lista_anos, lista_nascimentos):
  x = np.array([])
  for i in range(1,len(lista_anos) + 1): # Porque ta começando em 1 em vez de 0
    x = np.append(x,i)
  y = np.array(lista_nascimentos)
  n = x.size
  a = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2) - np.sum(x)**2)
  return a
```

```python
taxa_de_crescimento = calcula_taxa_de_crescimento(datas,valores)
taxa_de_crescimento
```

EQUAÇÂO PARA OBTER O COEFICIENTE LINEAR

b = Media(y) - â*Media(x)

```python
def calcula_coeficiente_linear(lista_anos, lista_nascimentos, taxa_de_crescimento):
  x = np.array([])
  for i in range(1,len(lista_anos) + 1): # Porque ta começando em 1 em vez de 0
    x = np.append(x,i)
  y = np.array(lista_nascimentos)
  a = taxa_de_crescimento
  b = np.mean(y) - a*np.mean(x)

  return b
```

```python
coef_linear = calcula_coeficiente_linear(datas, valores, taxa_de_crescimento)
coef_linear
```

```python
x = np.array([])
for i in range(1,len(datas) + 1): # Porque ta começando em 1 em vez de 0
  x = np.append(x,i)

regressao_linear_nascimentos = taxa_de_crescimento*x + coef_linear
regressao_linear_nascimentos
```
